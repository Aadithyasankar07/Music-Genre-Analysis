```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('../data/music.csv')

# Overview
print(df.head())
print(df.describe())
print(df.info())

# Most common genres
sns.countplot(data=df, y='genre', order=df['genre'].value_counts().index)
plt.title('Genre Frequency')
plt.show()

# Popularity by genre
sns.boxplot(data=df, x='genre', y='popularity')
plt.xticks(rotation=45)
plt.title('Popularity by Genre')
plt.show()

# Correlation heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Between Features')
plt.show()

# Danceability vs Energy (Interactive)
fig = px.scatter(df, x='danceability', y='energy', color='genre', hover_data=['track', 'artist'])
fig.show()
```
